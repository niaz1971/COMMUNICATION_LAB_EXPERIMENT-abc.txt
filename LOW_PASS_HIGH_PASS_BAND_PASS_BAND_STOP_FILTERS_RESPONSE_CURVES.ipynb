```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

# =============================================
# Step 1: Filter specifications
# =============================================
fs = 1000.0        # sampling frequency (Hz)
cutoff = 100.0     # low‑pass cutoff frequency (Hz)
order = 4          # filter order

# Normalize cutoff frequency (0 … 1, where 1 = fs/2)
nyquist = 0.5 * fs
normal_cutoff = cutoff / nyquist

# =============================================
# Step 2: Design digital low‑pass Butterworth filter
# =============================================
b, a = signal.butter(order, normal_cutoff, btype='low', analog=False)

# =============================================
# Step 3: Compute frequency response (magnitude only)
# =============================================
w, h = signal.freqz(b, a, worN=2000)

# Convert w from radians/sample to Hz
freq = w * fs / (2 * np.pi)

# Magnitude in dB
magnitude_dB = 20 * np.log10(np.abs(h))

# Plot only the magnitude (frequency) response
plt.figure(figsize=(8, 5))
plt.semilogx(freq, magnitude_dB, 'b', linewidth=2)
plt.axvline(cutoff, color='red', linestyle='--', label=f'cutoff = {cutoff} Hz')
plt.title('Low‑pass filter frequency response (magnitude)')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Magnitude [dB]')
plt.grid(True, which='both', axis='both')
plt.legend()
plt.xlim(10, fs/2)  # show from 10 Hz to Nyquist
plt.ylim(-80, 5)    # typical dB range
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

# =============================================
# Step 1: Filter specifications
# =============================================
fs = 1000.0              # sampling frequency (Hz)
f_low  = 50.0            # lower cutoff frequency (Hz)
f_high = 200.0           # upper cutoff frequency (Hz)
order  = 4               # filter order

# Normalize cutoff frequencies (0 … 1, where 1 = fs/2)
nyquist = 0.5 * fs
normal_cutoff = [f_low / nyquist, f_high / nyquist]   # band edges

# =============================================
# Step 2: Design digital band‑pass Butterworth filter
# =============================================
b, a = signal.butter(order, normal_cutoff, btype='band', analog=False)

# =============================================
# Step 3: Compute frequency response (magnitude only)
# =============================================
w, h = signal.freqz(b, a, worN=2000)

# Convert w from radians/sample to Hz
freq = w * fs / (2 * np.pi)

# Magnitude in dB
magnitude_dB = 20 * np.log10(np.abs(h))

# Plot only the magnitude (frequency) response
plt.figure(figsize=(8, 5))
plt.semilogx(freq, magnitude_dB, 'b', linewidth=2)
plt.axvline(f_low,  color='red', linestyle='--', label=f'f_low  = {f_low} Hz')
plt.axvline(f_high, color='green', linestyle='--', label=f'f_high = {f_high} Hz')
plt.title('Band‑pass filter frequency response (magnitude)')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Magnitude [dB]')
plt.grid(True, which='both', axis='both')
plt.legend()
plt.xlim(10, fs/2)  # show from 10 Hz to Nyquist
plt.ylim(-80, 5)    # typical dB range
plt.tight_layout()
plt.show()
```
