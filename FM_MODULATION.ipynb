```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
Ac = 1.0       # Carrier amplitude
Am = 1.0       # Modulating signal amplitude
fc = 40        # Carrier frequency (Hz)
fm = 4         # Modulating frequency (Hz)
beta = 6.0     # Modulation index (beta = delta_f / fm)

# Time vector
fs = 2000      # High sampling rate for smooth carrier waves
t = np.linspace(0, 0.5, int(fs * 0.5))

# Signals
m_t = Am * np.cos(2 * np.pi * fm * t)  # Message
c_t = Ac * np.cos(2 * np.pi * fc * t)  # Unmodulated Carrier

# FM formula: s(t) = Ac * cos(2 * pi * fc * t + beta * sin(2 * pi * fm * t))
# Because integral of cos(2*pi*fm*t) is sin(2*pi*fm*t) / (2*pi*fm)
s_t = Ac * np.cos(2 * np.pi * fc * t + beta * np.sin(2 * np.pi * fm * t))

# Plotting
fig, axes = plt.subplots(3, 1, figsize=(10, 8), sharex=True)

axes[0].plot(t, m_t, 'g', label='m(t) (Message)')
axes[0].set_title('Modulating Signal (Message) $m(t)$')
axes[0].set_ylabel('Amplitude')
axes[0].grid(True)

axes[1].plot(t, c_t, 'b', label='c(t) (Carrier)')
axes[1].set_title('Unmodulated Carrier Signal $c(t)$')
axes[1].set_ylabel('Amplitude')
axes[1].grid(True)

axes[2].plot(t, s_t, 'r', label='s(t) (FM Signal)')
axes[2].set_title('Frequency Modulated Signal $s(t)$')
axes[2].set_xlabel('Time (s)')
axes[2].set_ylabel('Amplitude')
axes[2].grid(True)

plt.tight_layout()
plt.savefig('frequency_modulation_plot.png', dpi=300)
print("FM plot saved successfully.")
```
