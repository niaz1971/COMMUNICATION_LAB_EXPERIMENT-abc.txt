```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
Am = 1.0       # Modulating signal amplitude
Ac = 1.0       # Carrier signal amplitude
fm = 5         # Modulating frequency (Hz)
fc = 50        # Carrier frequency (Hz)

# Time vector
fs = 1000      # Sampling frequency
t = np.linspace(0, 0.4, int(fs * 0.4))

# Signals
m_t = Am * np.cos(2 * np.pi * fm * t)  # Modulating signal
c_t = Ac * np.cos(2 * np.pi * fc * t)  # Carrier signal
s_t = m_t * c_t                        # DSB-SC AM Signal (m(t) * c(t))

# Plotting using subplots (no plt.figure())
fig, axes = plt.subplots(3, 1, figsize=(10, 8), sharex=True)

# Modulating Signal m(t)
axes[0].plot(t, m_t, 'g', label='m(t) (Message)')
axes[0].set_title('Modulating Signal $m(t)$')
axes[0].set_ylabel('Amplitude')
axes[0].grid(True)

# Carrier Signal c(t)
axes[1].plot(t, c_t, 'b', label='c(t) (Carrier)')
axes[1].set_title('Carrier Signal $c(t)$')
axes[1].set_ylabel('Amplitude')
axes[1].grid(True)

# Modulated Signal s(t) = m(t) * c(t)
axes[2].plot(t, s_t, 'r', label='s(t) = m(t) * c(t)')
axes[2].plot(t, m_t, 'k--', alpha=0.5, label='Envelope')
axes[2].plot(t, -m_t, 'k--', alpha=0.5)
axes[2].set_title('Modulated Signal $s(t) = m(t) \times c(t)$ (DSB-SC)')
axes[2].set_xlabel('Time (s)')
axes[2].set_ylabel('Amplitude')
axes[2].legend(loc='upper right')
axes[2].grid(True)

plt.tight_layout()
plt.savefig('dsb_sc_modulation_plot.png', dpi=300)
print("DSB-SC plot successfully saved.")
```
